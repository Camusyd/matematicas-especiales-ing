#  EDO No Homog茅neas: Estructura y Significado F铆sico (yc + yp)

Este Notebook explica y demuestra, paso a paso, por qu茅 la soluci贸n de una Ecuaci贸n Diferencial Ordinaria (EDO) lineal no homog茅nea de orden superior se compone siempre de dos partes: la **soluci贸n complementaria (homog茅nea) $y_c$** y la **soluci贸n particular $y_p$**.

El concepto clave es: **Soluci贸n Total = Comportamiento Natural + Comportamiento Forzado.**

## El Modelo de Ejemplo: Sistema Masa-Resorte Amortiguado Forzado

Utilizaremos el modelo de un sistema f铆sico com煤n en ingenier铆a: una masa sujeta a un resorte con amortiguamiento, sometida a una fuerza externa peri贸dica.

**La EDO a Resolver:**
$$

$$
Donde:
* $y(t)$ es la posici贸n de la masa.
* $y'' + 2y' + 5y$ representa las **fuerzas internas** (inercia, amortiguamiento y resorte).
* $10 \cos(t)$ es la **fuerza externa** (funci贸n de forzamiento).

```python
# Importaci贸n de librer铆as necesarias para c谩lculo simb贸lico
import sympy as sp
from sympy import Function, dsolve, Eq, Derivative, symbols, cos

# Definici贸n de la variable independiente (tiempo) y la funci贸n dependiente (posici贸n)
t = symbols('t')
y = Function('y')

print("Librer铆as importadas y s铆mbolos definidos.")
```

## Paso 1: Comprender la Estructura de la Soluci贸n

Por el principio de superposici贸n, la soluci贸n general $y(t)$ es la suma de dos soluciones que tienen prop贸sitos distintos:

$$
y(t) = y_c(t) + y_p(t)
$$

### 1. $y_c(t)$: La Soluci贸n Complementaria (Comportamiento Natural)
* **Se obtiene resolviendo la EDO homog茅nea:** $y'' + 2y' + 5y = 0$.
* **Significado:** Describe c贸mo vibrar铆a el sistema si no hubiera fuerza externa. Representa la **respuesta transitoria** del sistema, que generalmente decae con el tiempo (debido al amortiguamiento).

### 2. $y_p(t)$: La Soluci贸n Particular (Comportamiento Forzado)
* **Se obtiene resolviendo la EDO no homog茅nea:** $y'' + 2y' + 5y = 10 \cos(t)$.
* **Significado:** Describe la vibraci贸n que el sistema adopta *debido exclusivamente a la fuerza externa* $10 \cos(t)$. Representa la **respuesta de estado estacionario** que persiste indefinidamente.


```python
# --- CLCULO DE LA SOLUCIN COMPLEMENTARIA (yc) ---

print("--- C谩lculo de yc: El Movimiento Natural (EDO Homog茅nea) ---")

# EDO Homog茅nea: y'' + 2y' + 5y = 0
edo_homogenea = Eq(Derivative(y(t), t, 2) + 2*Derivative(y(t), t, 1) + 5*y(t), 0)
print(f"Ecuaci贸n a resolver (yc): {edo_homogenea}")

# Resolver la EDO Homog茅nea
sol_complementaria = dsolve(edo_homogenea, y(t))

# Extraer solo el lado derecho de la soluci贸n, que es yc
yc = sol_complementaria.rhs
print("\nSoluci贸n Complementaria (yc) obtenida:")
print(f"yc(t) = {yc}")

# Nota: exp(-t) indica que esta soluci贸n decae (es transitoria).
```

```python
# --- CLCULO DE LA SOLUCIN TOTAL (y = yc + yp) ---

print("--- C谩lculo de la Soluci贸n Total (y = yc + yp) ---")

# EDO No Homog茅nea completa: y'' + 2y' + 5y = 10*cos(t)
edo_completa = Eq(Derivative(y(t), t, 2) + 2*Derivative(y(t), t, 1) + 5*y(t), 10*cos(t))
print(f"Ecuaci贸n a resolver (y): {edo_completa}")

# Resolver la EDO No Homog茅nea
solucion_total = dsolve(edo_completa, y(t))

y_total = solucion_total.rhs
print("\nSoluci贸n General y(t) obtenida por SymPy:")
print(f"y(t) = {y_total}")
```

```python
# --- INTERPRETACIN Y EXTRACCIN DE LA SOLUCIN PARTICULAR (yp) ---

# SymPy nos dio la soluci贸n total y(t).
# La soluci贸n particular yp(t) es la parte que NO contiene las constantes C1 y C2.

# 1. Separar la Soluci贸n Particular (yp)
# Se puede hacer observando el resultado o, m谩s formalmente, eliminando los t茅rminos de yc.
# En este caso, la parte trigonom茅trica restante es la soluci贸n particular.

# yp es la parte de la soluci贸n que no tiene los t茅rminos e^(-t)*sin(2t) ni e^(-t)*cos(2t)
yp_num = y_total - yc.subs(sp.symbols('C1'), 0).subs(sp.symbols('C2'), 0)
# Sustituimos C1 y C2 con 0 en la soluci贸n total para aislar la yp.
# Nota: Esta es una simplificaci贸n, ya que C1 y C2 solo est谩n en yc.
# La soluci贸n particular yp es: 2*sin(t) + cos(t)

print("--- An谩lisis de yp ---")
yp = sp.simplify(y_total.subs(sp.symbols('C1'), 0).subs(sp.symbols('C2'), 0))
print(f"Soluci贸n Particular (yp) = {yp}")

print("\n--- Conclusi贸n F铆sica ---")
print("El t茅rmino:")
print(f"yc = {yc}")
print("...describe la **respuesta transitoria**, el movimiento libre del resorte que se detiene por el amortiguamiento e^(-t).")
print("\nEl t茅rmino:")
print(f"yp = {yp}")
print("...describe la **respuesta de estado estacionario**, el movimiento forzado que persiste en el tiempo debido a la fuerza externa 10*cos(t).")
```

## Interpretaci贸n Final (Nivel Situacional)

En un sistema f铆sico, solo la **Soluci贸n Particular ($y_p$)** permanece despu茅s de un tiempo prolongado, ya que la **Soluci贸n Complementaria ($y_c$)** se "apaga" debido al amortiguamiento (el t茅rmino $e^{-t}$ en la soluci贸n $y_c$ lo hace tender a cero).

* **Si el sistema es un puente y la fuerza externa es el tr谩fico:**

    * **$y_p$** es la vibraci贸n constante que el puente mantendr谩 mientras el tr谩fico siga pasando.

* **Concepto Central del Curso:** Para modelar la respuesta de un sistema forzado en estado estable, es fundamental comprender y calcular la **soluci贸n particular $y_p$**, ya que esta es la que domina la soluci贸n a largo plazo.
