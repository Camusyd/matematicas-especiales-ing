##  El Principio de la Sustituci贸n

Toda EDO lineal de orden $n$ puede escribirse en la forma:
$$
\mathbf{y^{(n)} = f(t, y, y', y'', \dots, y^{(n-1)})}
$$

Para transformarla, introducimos **$n$ nuevas variables**, donde cada nueva variable es la derivada de la anterior.



## Ejemplo F谩cil

Convertiremos una EDO de segundo orden, que modela un oscilador amortiguado sin forzamiento (homog茅neo):
$$
\mathbf{y'' + 4y' + 3y = 0}
$$

```python
# Importaci贸n de librer铆as necesarias
import sympy as sp
from sympy import Function, dsolve, Eq, Derivative, symbols, Matrix

# Definici贸n de la variable independiente (tiempo)
t = symbols('t')

# Definici贸n de la funci贸n original (y) y las nuevas variables (x1, x2)
y = Function('y')
x1 = Function('x1')(t)
x2 = Function('x2')(t)

print("S铆mbolos y variables definidas.")
```

## Paso 1: Definici贸n de Variables

Introducimos dos nuevas variables dependientes, $\mathbf{x_1}$ y $\mathbf{x_2}$, ya que la EDO es de segundo orden ($n=2$):

1.  **Variable original:** La primera variable es la funci贸n original.
    $$
    \mathbf{x_1(t) = y(t)}
    $$

2.  **Primera Derivada:** La segunda variable es la primera derivada de $y$.
    $$
    \mathbf{x_2(t) = y'(t)}
    $$

## Paso 2: Crear el Sistema de Primer Orden

Ahora, derivamos cada una de las nuevas variables. La meta es expresar $\mathbf{x'_1}$ y $\mathbf{x'_2}$ 煤nicamente en t茅rminos de $x_1$ y $x_2$.

### Ecuaci贸n 1: La Definici贸n
Derivamos $\mathbf{x_1}$:
$$
\mathbf{x'_1} = \frac{d}{dt}(y) = y'
$$
Por la definici贸n del Paso 1, sabemos que $y' = x_2$.
$$
\mathbf{x'_1 = x_2} \quad \text{(Ecuaci贸n de Definici贸n)}
$$

### Ecuaci贸n 2: La EDO Sustituida
Derivamos $\mathbf{x_2}$:
$$
\mathbf{x'_2} = \frac{d}{dt}(y') = y''
$$
Despejamos $y''$ de la EDO original ($y'' + 4y' + 3y = 0$):
$$
y'' = -4y' - 3y
$$
Sustituimos $y$ por $x_1$ y $y'$ por $x_2$:
$$
\mathbf{x'_2 = -3x_1 - 4x_2} \quad \text{(Ecuaci贸n del Sistema)}
$$

##  El Sistema Acoplado de EDOs

Hemos transformado la EDO de segundo orden en un sistema de dos EDOs de primer orden acopladas:

$$
\begin{cases} \mathbf{x'_1 = x_2} \\ \mathbf{x'_2 = -3x_1 - 4x_2} \end{cases}
$$
Este sistema es **lineal y homog茅neo** y puede resolverse utilizando matrices.

```python
# --- Representaci贸n Matricial del Sistema ---

# El sistema de la forma X' = A * X
# donde X = [x1, x2] y X' = [x1', x2']

# 1. Definir la Matriz de Coeficientes A:
# Fila 1 (x1'): [coeficiente de x1, coeficiente de x2] -> [0, 1]
# Fila 2 (x2'): [coeficiente de x1, coeficiente de x2] -> [-3, -4]

A = Matrix([[0, 1], [-3, -4]])

print("--- Matriz del Sistema A ---")
print(A)

# 2. Vector de Estado X
X = Matrix([[x1], [x2]])
print("\n--- Vector de Estado X ---")
print(X)

# La ecuaci贸n del sistema es: X' = A * X
# Esta forma matricial permite aplicar el c谩lculo de Eigenvalores y Eigenvectores
# para resolver el sistema y analizar su estabilidad (Retratos de Fase).
```
