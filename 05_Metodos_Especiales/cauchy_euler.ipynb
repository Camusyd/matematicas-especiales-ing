##  Objetivo

Comprender y aplicar el **M茅todo de Cauchy-Euler** (tambi茅n conocidas como Ecuaciones Equidimensionales), un tipo especial de Ecuaci贸n Diferencial Ordinaria (EDO) con coeficientes variables que puede transformarse en una EDO de coeficientes constantes.

##  Caracter铆stica Principal

Una EDO es de tipo Cauchy-Euler si puede escribirse en la forma:
$$
\mathbf{a_n x^n \frac{d^n y}{dx^n} + a_{n-1} x^{n-1} \frac{d^{n-1} y}{dx^{n-1}} + \dots + a_1 x \frac{dy}{dx} + a_0 y = g(x)}
$$
Donde la potencia de $x$ siempre **coincide con el orden de la derivada** de $y$ a la que multiplica (por ejemplo, $x^2$ multiplica a $y''$).

---

##  Principio del M茅todo (El "Truco")

El m茅todo asume que la soluci贸n de la parte homog茅nea tiene una forma de **potencia**:
$$
\mathbf{y = x^m}
$$
Al sustituir esta forma en la EDO, se reduce el problema diferencial a la resoluci贸n de un **polinomio algebraico** (la Ecuaci贸n Caracter铆stica de Cauchy-Euler), similar al m茅todo de coeficientes constantes.

## Ejemplo F谩cil

Resolveremos la siguiente EDO homog茅nea de segundo orden:
$$
\mathbf{x^2 y'' - 2x y' + 2y = 0}
$$

```python
# Importaci贸n de librer铆as necesarias para c谩lculo simb贸lico
import sympy as sp
from sympy import Function, dsolve, Eq, Derivative, symbols, log

# Definici贸n de la variable independiente (x) y la funci贸n dependiente (y)
x = symbols('x')
y = Function('y')

print("Librer铆as importadas y s铆mbolos definidos.")
```

## Paso 1: Establecer la Ecuaci贸n Caracter铆stica

Sustituimos la soluci贸n propuesta $y = x^m$ y sus derivadas ($y' = mx^{m-1}$, $y'' = m(m-1)x^{m-2}$) en la EDO:

$$
x^2 [m(m-1)x^{m-2}] - 2x [mx^{m-1}] + 2 [x^m] = 0
$$

Simplificando (multiplicando las potencias de $x$):
$$
m(m-1)x^m - 2m x^m + 2 x^m = 0
$$

Factorizamos $x^m$:
$$
x^m [m(m-1) - 2m + 2] = 0
$$

La **Ecuaci贸n Caracter铆stica de Cauchy-Euler** es el polinomio dentro del corchete (asumiendo $x \neq 0$):
$$
m^2 - m - 2m + 2 = 0
$$
$$
\mathbf{m^2 - 3m + 2 = 0}

```python
# Definici贸n del s铆mbolo para la ra铆z 'm'
m = symbols('m')

# Definici贸n de la Ecuaci贸n Caracter铆stica (Polinomio)
ecuacion_caracteristica = Eq(m**2 - 3*m + 2, 0)
print(f"Ecuaci贸n Caracter铆stica: {ecuacion_caracteristica}")

# Resolver el polinomio para encontrar las ra铆ces 'm'
raices_m = sp.solve(ecuacion_caracteristica, m)

print(f"Ra铆ces obtenidas (m1, m2): {raices_m}")
```

## Paso 2: Construir la Soluci贸n General ($y_c$)

Existen tres casos posibles para las ra铆ces $m_1$ y $m_2$ que determinan la forma final de la soluci贸n:

### Caso 1: Ra铆ces Reales y Distintas ($m_1 \neq m_2$)
$$
\mathbf{y(x) = C_1 x^{m_1} + C_2 x^{m_2}}
$$
(Este es el caso que obtuvimos en nuestro ejemplo: $m_1 = 2$ y $m_2 = 1$).

### Caso 2: Ra铆ces Reales Repetidas ($m_1 = m_2 = m$)
$$
\mathbf{y(x) = C_1 x^{m} + C_2 x^{m} \ln(|x|)}
$$

### Caso 3: Ra铆ces Complejas Conjugadas ($m = \alpha \pm i\beta$)
$$
\mathbf{y(x) = x^{\alpha} [C_1 \cos(\beta \ln|x|) + C_2 \sin(\beta \ln|x|)]}
$$

```python
# Aplicando el Caso 1 (Ra铆ces Reales y Distintas)
m1 = raices_m[0]  # m1 = 2
m2 = raices_m[1]  # m2 = 1

# Construcci贸n de la soluci贸n general
C1, C2 = symbols('C1 C2')
y_solucion = C1 * x**m1 + C2 * x**m2

print("--- Soluci贸n Final ---")
print(f"La Soluci贸n General es:")
print(f"y(x) = {y_solucion}")

# --- Comprobaci贸n con dsolve de SymPy ---
print("\n--- Comprobaci贸n (SymPy dsolve) ---")
edo = Eq(x**2 * Derivative(y(x), x, 2) - 2*x * Derivative(y(x), x, 1) + 2*y(x), 0)
sol_dsolve = dsolve(edo, y(x))
print(f"Resultado de dsolve: {sol_dsolve.rhs}")
```

##  Conclusi贸n

La soluci贸n general de la EDO $x^2 y'' - 2x y' + 2y = 0$ es:
$$
\mathbf{y(x) = C_1 x^2 + C_2 x}
$$
El m茅todo de Cauchy-Euler nos permiti贸 transformar una EDO compleja con coeficientes variables en un simple polinomio cuadr谩tico, ilustrando la potencia de las t茅cnicas anal铆ticas para simplificar el proceso de resoluci贸n.
